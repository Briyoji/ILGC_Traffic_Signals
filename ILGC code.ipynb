```python
import math
print("Hello System!")
def main():
    try:
        D=0
        d=1
        lat_sig,long_sig = input1_pos_sig()
        while (D<=d):
            pos_car_lat , pos_car_long=input2_pos_car()
            print()
            d     = distance3(lat_sig,long_sig,pos_car_lat,pos_car_long)
            D = d            
            if d==0:
                print("(Temp.) you reached at signal")
                main()
            if (d <=1000):    
                T,c_speed = input4_c_speed(d)
                rem_t = input5_s_cycle(T)
                Ex_t  = input6_opt_fast(rem_t,T)
                r=required_speed7(d,Ex_t)
                if c_speed==r:
                    print("Your pace is good, keep on going!")
             
    except ValueError:
        print("Value is not in correct format")
        
def input1_pos_sig():
    lat_sig = float(input("Enter upcoming signal's lat: "))
    long_sig=float(input("Enter upcoming signal's long: "))
    return lat_sig,long_sig
def input2_pos_car():
    pos_car_lat= float(input("car lat:"))
    pos_car_long=float(input("car long: "))
    return pos_car_lat , pos_car_long        
def distance3(lat_sig,long_sig,pos_car_lat,pos_car_long):
    d=math.sqrt(((lat_sig - pos_car_lat)**2 + (long_sig - pos_car_long)**2))
    print("Distance need to cover: ", d,"meter")
    return d

def input4_c_speed(d):
    try:
        c_speed=float(input("Current speed in meter/sec: "))
        T=d/c_speed
        #print("remaining time to cover present distance",T,"sec")
        return T,c_speed
    except ZeroDivisionError:
        print("Your current speed is zero now. Thanks!")
def input5_s_cycle(T):
    s_cycle=float(input("Enter the cycle of signal in seconds: "))
    rem_t=T%s_cycle
    #print("Remaining time for red signal",rem_t)
    return rem_t
def input6_opt_fast(rem_t,T):
    opt_fast=int(input("For optimum press 1, For fast press 2: "))
    if opt_fast==1:
        Ex_T=T+rem_t
    elif opt_fast==2:
        Ex_T=T-rem_t
    else:
        print("You did not choose correct number")
    #print("New time w.r.t signal cycle",Ex_T)
    return Ex_T
def required_speed7(d,Ex_T):
    try:
        Ex_speed=d/Ex_T
        print()
        print("Required speed to catch next green signal",Ex_speed,"m/sec")
        print()
        return Ex_speed
    except ZeroDivisionError:
        print("Your distance is zero now. Please be at your own pace. Thanks!")
        
        
        
main()
```
